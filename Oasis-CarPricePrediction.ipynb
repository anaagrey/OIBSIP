```python
import pandas as pd
import numpy as np
!pip install scikit-learn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
dataset= pd.read_csv('car data.csv')
```

```python
dataset.columns
```

```python
le=LabelEncoder()
dataset['New_Fuel']=le.fit_transform(dataset['Fuel_Type'])
dataset['New_Sell']=le.fit_transform(dataset['Selling_type'])
dataset['New_Transmission']=le.fit_transform(dataset['Transmission'])
```

```python
dataset.head()
```

```python
dataset['Year']=dataset['Year'].subtract(2024)
```

```python
dataset.sample()
```

```python
X=dataset.drop(columns=['Fuel_Type','Selling_type','Transmission','Car_Name'])
X.drop(columns='Selling_Price',inplace=True)
Y=dataset['Selling_Price']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import accuracy_score,mean_absolute_error
error=mean_absolute_error(Y_test,Y_pred)
print(error)
```
